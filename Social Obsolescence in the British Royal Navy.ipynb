## Introduction

The emergence of new technologies and the obsolescence of old technologies are two sides of the same coin. 19th-century Britain, a highly advanced industrial society, found itself in the era of "Social Obsolescence," as pointed out by Karl Marx(citation). This phenomenon was evident in various scenarios, where the value of certain items in society diminished due to factors such as the introduction of new products and a decrease in economic efficiency, even though they were still materially usable.

During this era of social obsolescence, slogans such as efficiency and retrenchment were prominent. Pursuit of social and economic efficiency had a strong impact, and public finance played a deep role in the lives of citizens. The proper utilization of national taxes was a significant concern for society, leading to scrutiny of government spending through media, parliament, and special committees(citation). Of particular interest in this paper is the concept of "economies of waste" within the British Royal Navy. The emergence of floating batteries during the Crimean War highlighted the inadequacy of wooden ships' defensive capabilities in actual combat. This realization led to the construction of armored ships like Napoleon III's French Navy's Gloire and subsequently the British Navy's HMS Warrior(citation).

Sir John Fisher, who became the First Sea Lord in 1905, pursued military and financial efficiency by several approaches, such as personnel and education reforms within the Navy, the construction of modern ships and the decommissioning of obsolete ones. The chief result of his reforms was the construction of the Dreadnought-class battleship in 1905, and thus marking many ships as obsolete(<cite data-cite="3597926/D5ZCUUZC"></cite>).

While previous studies discuss the disposal of old ships in the British Navy, they primarily focus on the Fisher Reforms era. Moreover, these explain mainly from perspective of naval history, suggesting that the disposal of old warships aimed to reduce maintenance costs, generate funds for refreshing the Navy's personnel system, or efficiently utilize harbor facilities for new ships(citation).

However, what happened next after disposal? Massive resources like naval ships couldn't simply disappear. During the rapid military technological innovation period following the Crimean War, masses of resources composed of wood, iron, and steel were rendered redundant. Thus, the concept of reusing and recycling these resources was emerging. Concepts such as the reuse of human waste in London or Marx's late-life considerations regarding soil lifecycle recycling were emerging during this time(citation).

Paying attention to actors outside the British Navy, a different historical perspective arises. Examining the Japanese Navy during the late Edo and Meiji periods, there was a high demand for Western-style vessels, and it is well-known that former British Navy ships had second careers as active or training ships(citation). Beyond reusing ships, dismantling ship hulls to obtain resources such as wood and iron for recycling into the steel industry or military manufacturing was observed(citation).

This 'sustainable', in today's sense, structure is intriguingly suggested by an oil painting featured on the refreshed £20 banknote in the United Kingdom in 2020. The painting, Joseph Mallord William Turner's 'The Fighting Temeraire tugged to her last berth to be broken up' as shown in Figure 1, vividly portrays the scene of the ship, a former British Navy vessel, being towed to the dismantling yard of Castle, a specialist in the deconstruction of wooden vessels in the 19th century.

According to Leo Costello, art historian,

> Even the name contains a sense of this dual movement: "berth" puns with "birth" so that decay and growth are equally implied here. ...The metaphor of the ship in the *Temeraire* is not a means of bringing together a set of prior, disparate parts, but rather of implying a future breaking up, back into parts (<cite data-cite="3597926/D5YTE9JN"></cite>).

In essence, this suggests that the outdated vessels of the British Navy, no longer of use to the British Navy itself, held value for actors outside the British Navy. Frank Trentmann, historian in consumption, theorizes that consumption in societies should be seen not as a linear process of production, consumption, and disposal, but as a cyclic process of production, consumption, disposal, reuse, and re-consumption (<cite data-cite="3597926/QVRP2MTS"></cite>). This implies that objects embark on a second or even third career after being discarded.

The goal of this article is to unveil the fundamental facts about the disposal of obsolete ships in the British Navy, from 1860s to 1920s. It aims to clarify which types of vessels were disposed, when such disposal was significantly prioritized, and how it was a product of naval policy. This article sheds light on the old ships from various perspectives including naval policy, national finance, diplomatic strategy, and industrial aspects, within the space constraints. To illustrate the comprehensive overview of the disposal of obsolete ships within the British Navy, it is vital to understand the naval vessel information as thoroughly as possible. However, an extensive amount of naval vessel data is scattered across various types of primary sources and sometimes involves conflicting data that requires scrutiny and selection. This makes it deemed impossible for a single researcher to construct the accurate dataset. The task at hand is not to compile an exhaustive and complete database of outdated ships from the vast sea of historical records. Rather, the aim is to compare and evaluate multiple reference books that have been used in British naval history research, assess their editing approaches and the strengths and weaknesses of their datasets, and by combining useful data for unveiling the aspects of outdated ship disposal, establish a unique database.

Among the datasets documenting the careers of British Navy vessels from 1850 to 1930, the four under consideration are as follows: A) Volumes 1 and 2 of the *Conway's All the World's Fighting Ships* series, B) *The Sail & Steam Navy List* by David Lyon and Rif Winfield, C) *British Warships in the Age of Sail 1817–1863* by Rif Winfield (which succeeded the previous work), and D) the third edition of *Ships of the Royal Navy* by Ben Warlow, an expanded revision of J.J. Colledge's work.

The series in A) comprehensively compiles naval vessel information from over 40 countries worldwide, from the mid-19th to the late 20th centuries. They explain about the technical characteristics of different eras and types of vessels. Its coverage beyond major navies makes it valuable for international naval strength comparisons. However, the reliability of certain information is restricted to the involvement of leading scholars in editing, and the preface's assertion that 'it is impossible to enumerate all information sources' cannot be underestimated.

B) and C) deal with the data of British Navy vessels from the era of sail ships. Volume 4 of C) covers the period 1817–1863. References to the naval records, encompassing information from construction to repair and disposal, are used in these works to provide detailed descriptions of individual vessel careers. However, it is important to note that not all the time periods covered by this paper are included.

D) contains data on 18,000 British Navy vessels from the 15th to the early 21st centuries. The provided data includes vessel names, types, ranks, classes, builders, completion years, dimensions, armaments, armor, tonnages, propulsion sources, hull materials, career summaries, and fates after disposal. The primary sources are ADM 180 and the Navy Lists published quarterly by the government since 1815. While not as detailed as B) and C), the data in these lists about vessel careers is useful for getting an overview. Though, some areas suffer from a very high data loss rate, demanding caution in analysis.

Given the analysis of the datasets, this paper mainly relies on D) since it covers the entire analysis period. References to A)–C) are made as needed to supplement missing data. When incorporating naval vessel information into the database, the following principles were adhered to: (1) the year of ownership loss for the British Navy falls within the range of 1850 to 1930, (2) the vessel is not owned by the British colonial navy, (3) it was completed (excluding vessels whose construction was halted or cancelled), and (4) it served a mission in the British Navy. Consequently, a database of approximately 4,500 naval vessels was constructed.

## Hermeneutic strategy 

When utilizing the information included in the dataset mentioned above, a descriptive statistical analysis using the programming language Python was conducted. The primary libraries employed were pandas for aiding in statistical analysis, and matplotlib and seaborn for data visualization.

The backdrop for those descriptive statistical analyses is rooted in Tim Hitchcock's 'data-first' approach. In essence, this approach aims to intentionally present a comprehensive dataset and lead readers to specific conclusions (<cite data-cite="3597926/HEYSZR3R"></cite>). With the broad dissemination of digital archives, the historical data available to us has exponentially grown. Thus, to elucidate the reason why specific aspects were focused on and their validity, quantitative methods such as text mining, GIS, and statistics contribute to exploratory data analysis. This entails temporarily distancing yourself from the established foundations of prior research and personal interests, and employing quantitative methods exploratorily to identify noteworthy points based on some form of measurement obtained(<cite data-cite="3597926/MIPC49B8"></cite>, <cite data-cite="3597926/GX6HCW5G"></cite>). This process demonstrates the rationale behind the selection to third parties. It is important to note here that the results of data analysis should not be accepted uncritically; rather, one should ensure that important information is not overlooked during the analysis process. The crucial aspect is to iterate between exploratory data analysis in the preliminary stages and in-depth reading in subsequent stages to deepen the examination.

This paper proceeds from the results of data visualization based on descriptive statistical analysis to delve into specific noteworthy cases. It adopts a narrative strategy that guides readers from the overall perspective to individual cases, a structure that is conscious of transitioning from distant reading to close reading.
